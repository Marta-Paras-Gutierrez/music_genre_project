## 8.- Selección de variables clave

Como se ha podido observar, el dataset en bruto contiene múltiples variables acústicas extraídas mediante técnicas de procesamiento digital de audio. Debido a este elevado número de columnas, se ha realizado un proceso de filtrado estratégico para seleccionar únicamente aquellas variables que aportan la mayor relevancia interpretativa y visual para nuestro análisis exploratorio, evitando así la redundancia de datos.

Durante el estudio nos centraremos en los siguientes parámetros fundamentales, que son los más utilizados en las tareas profesionales de clasificación automática y análisis musical:

- **tempo**: Mide el ritmo de la canción (en BPM), permitiendo estudiar la velocidad característica de cada género.

- **rms_mean**: Representa el valor medio de la raíz cuadrada media de la señal, una métrica clave para medir la energía y el volumen percibido del audio.

- **spectral_centroid_mean** y **spectral_bandwidth_mean**: Indican el centro de gravedad y el ancho de banda del espectro de frecuencias. Nos ayudan a interpretar el brillo espectral y la forma del sonido (si la música es más "oscura" o más "brillante").

- **zero_crossing_rate_mean**: Mide la tasa de cruces por cero de la señal, una variable muy útil para analizar la complejidad sonora, la presencia de ruido o cambios abruptos en el audio (como en las guitarras distorsionadas o elementos percusivos).

- **chroma_stft_mean**: Analiza las características tonales proyectando el espectro en las 12 notas de la escala cromática, ideal para capturar la estructura armónica de la música.

- **Coeficientes MFCC** (mfcc_mean): Coeficientes cepstrales en las frecuencias de Mel, fundamentales para obtener una representación frecuencial precisa del timbre del audio, simulando cómo el oído humano percibe el sonido.

- **label**: La variable categórica que contiene el género musical y que actuará como nuestro eje de clasificación.

## 10.- Creación de nuevas variables

Con el objetivo de enriquecer el análisis exploratorio, se ha diseñado una fase de ingeniería de variables (feature engineering) para generar nuevas métricas derivadas a partir de las características acústicas que ya existen en el dataset en bruto.

Este proceso nos permite resumir información compleja y combinar distintas dimensiones del sonido en métricas mucho más interpretables, facilitando la detección de patrones acústicos que no se aprecian a simple vista.

Las nuevas variables que se incorporan al pipeline de datos son:

- **song_complexity** (Complejidad sonora): Una medida aproximada que combina variables de cambio de frecuencia y ritmo para evaluar qué tan saturada o compleja es la estructura de la señal de audio.

- **mfcc_mean_total** (Promedio global de MFCC): Un indicador que promedia el comportamiento medio de los coeficientes MFCC para resumir el timbre general de la canción en un solo valor fácilmente comparable.

- **energy_tempo_ratio** (Relación energía/ritmo): Un ratio que calcula la relación entre la energía percibida (volumen/potencia) y el tempo (BPM), ideal para mapear de forma visual qué géneros son rápidos pero acústicamente ligeros frente a los que son lentos pero muy densos.

```python
# Feature 1: Complejidad sonora basada en variables espectrales y ritmo

df_30["song_complexity"] = (
    df_30["spectral_bandwidth_mean"]
    + df_30["zero_crossing_rate_mean"]
    + df_30["tempo"]
) / 3
```

```python
# Feature 2: Media global con selección automática de coeficientes MFCC

mfcc_cols = [
    col for col in df_30.columns
    if "mfcc" in col and "mean" in col
]

# Cálculo de media MFCC
df_30["mfcc_mean_total"] = df_30[mfcc_cols].mean(axis=1)
```

```python
# Feature 3: Relación energía y ritmo (velocidad de la canción)

df_30["energy_tempo_ratio"] = (
    df_30["rms_mean"] / df_30["tempo"]
)
```

```python
# Visualización de las nuevas variables generadas

df_30[
    [
        "song_complexity",
        "mfcc_mean_total",
        "energy_tempo_ratio"
    ]
].head()
```

Las nuevas variables generadas permiten representar de forma más compacta determinadas propiedades acústicas de las canciones.

Estas features podrán ayudar a:

- identificar patrones entre géneros musicales

- simplificar ciertas relaciones acústicas

- mejorar futuros modelos de clasificación automática

Esto se debe a que las ventanas temporales más pequeñas capturan fluctuaciones locales del audio con mayor sensibilidad —como un puente musical lento, un solo de batería acelerado o un silencio momentáneo—, mientras que las muestras de canciones completas de 30 segundos (en tono verde) suavizan estos picos, generando métricas más estables, homogéneas y promediadas.
